# 1. Introduction

```julia
using Random
using Plots
using Statistics
using LinearAlgebra
```

```julia
function f(x::Number)
    sin(2π*x)
end

function create_data(func, sample_size, std)
    x = LinRange(0, 1, sample_size)
    t = f.(x) + randn(size(x))*std
    x, t
end

x_train, y_train = create_data(f, 10, 0.25)
x_test = LinRange(0, 1, 100)
y_test = f.(x_test)

plot(x_test, y_test)
plot!(x_train, y_train, seriestype=:scatter)
```

# Linear regression

```julia
function polyexpand(x, degree::Integer)
    coefs = Array{Float64}[]
    for n in 1:degree
        push!(coefs,x.^n)
    end
    push!(coefs,ones(size(x)))
    
    hcat(coefs...)
end


function polyfit_ls(x, y::Array{Float64,1}, degree::Integer)   
    xpn = polyexpand(x, degree)
    β = (xpn'*xpn)\xpn'*y
    
    β
end
```

```julia
plots = Any[]

for n in [1,3,5,9]
    β = polyfit_ls(x_train, y_train, n)
    pp = plot(x_train, y_train, seriestype=:scatter, label="training data")
    xpn_test = polyexpand(x_test, n)
    pp = plot!(x_test, xpn_test*β, label="fitting")
    pp = plot!(x_test, y_test, label="sin(2Pix)")
    push!(plots,pp)
end

plot(plots...,layout=(2,2),size=(800,600))
```

```julia
function rmse(errs)
    return sqrt(mean(errs.^2.))
end

train_errs = Number[]
test_errs = Number[]

for n in 1:9
    β = polyfit_ls(x_train, y_train, n)
    xpn_train = polyexpand(x_train, n)
    y_est = xpn_train*β
    train_err = y_train - y_est
    push!(train_errs,rmse(train_err))
    test_err = f.(LinRange(0, 1, 10)) - y_est
    push!(test_errs,rmse(test_err))
end

plot(train_errs, label="train error")
plot!(test_errs, label="test error")
```

# Ridge regression
\begin{equation}
\beta = (X^TX+ \lambda I )^{-1}X^T Y.
\end{equation}

```julia
function polyfit_ridge(x, y::Array{Float64,1}, λ::Float64, degree::Integer)   
    xpn = polyexpand(x, degree)
    β = (xpn'*xpn + λ.*LinearAlgebra.I)\xpn'*y
    
    β
end
```

```julia
plots = Any[]

for n in [1,3,5,9]
    β = polyfit_ridge(x_train, y_train, 1e-3, n)
    pp = plot(x_train, y_train, seriestype=:scatter, label="training data")
    xpn_test = polyexpand(x_test, n)
    pp = plot!(x_test, xpn_test*β, label="fitting")
    pp = plot!(x_test, y_test, label="sin(2Pix)")
    push!(plots,pp)
end

plot(plots...,layout=(2,2),size=(800, 600))
```

```julia
train_errs = Number[]
test_errs = Number[]

for n in 1:9
    β = polyfit_ridge(x_train, y_train, 1e-3, n)
    xpn_train = polyexpand(x_train, n)
    y_est = xpn_train*β
    train_err = y_train - y_est
    push!(train_errs,rmse(train_err))
    test_err = f.(LinRange(0, 1, 10)) - y_est
    push!(test_errs,rmse(test_err))
end

plot(train_errs, label="train error")
plot!(test_errs, label="test error")
```

# Bayesian regression

```julia
n=9
X = polyexpand(x_train, n)
X_test = polyexpand(x_test, n)
t = y_train

α=2e-3
β=2
w_mean=nothing
w_precision=nothing

function getprior(ndim::Integer)
    if w_mean!=nothing && w_precision!=nothing
        return w_mean, w_precision
    else
        return zeros(ndim), α.*Diagonal(ones(ndim))
    end
end

mean_prev, precision_prev = getprior(10)

w_precision = precision_prev .+ β * X'*X
w_mean = w_precision\(precision_prev*mean_prev .+ β*X'*t)
w_cov = inv(w_precision)
```

```julia
y = X_test*w_mean
y_std = sqrt.(sum((X_test*w_cov)'*X_test,dims=1))

plot(x_test,y_test,label="sin(2pi*x)")
plot!(x_test,y,ribbon=y_std,label="mean, std")
plot!(x_train,y_train,seriestype=:scatter,label="training data")
```

```julia
y = X_test*w_mean
y_std = sqrt.(sum((X_test*w_cov)'*X_test,dims=1))

plot(x_test, y_test, label="sin(2pi*x)")
plot!(x_test,y,ribbon=y_std, label="mean, std")
plot!(x_train, y_train, seriestype=:scatter, label="training data")
```
