### 2.1.1 The beta distributions

```julia
x = LinRange(0, 1, 100)

#plot(rand(100, 4), layout = 4)

l = @layout [a b; c d]
p1 = plot(Beta(0.1, 0.1), label="a=0.1 b=0.1")
p2 = plot(Beta(1., 1.), label="a=1 b=1")
p3 = plot(Beta(2., 3.), label="a=2 b=3")
p4 = plot(Beta(8., 4.), label="a=8 b=4")

plot(p1, p2, p3, p4, layout = l)
```

```julia
l = @layout [a; b]

beta0 = Beta(2, 2)
p1 = plot(beta0, label="Prior")

beta1 = Beta(3,2)
p2 = plot(beta1, label="Posterior")


plot(p1, p2, layout = l)
```

### 2.3 The Gaussian Distribution

```julia
l = @layout [a; b; c]

uniform = Uniform(0, 1)

d1 = rand(uniform, 100000)
p1 = histogram(d1, bins=50, label="N=1")

d2 = rand(uniform, 100000) + rand(uniform, 100000)
p2 = histogram(d2, bins=50, label="N=2")

d3 = rand(uniform, 100000)
for _ = 1:10
    d3 += rand(uniform, 100000)
end

p3 = histogram(d3, bins=50, label="N=10")

plot(p1, p2, p3, layout = l)
```

### 2.3.4 Maximum Likelihood for the Gaussian

```julia
norm = Normal(1,2)
x = reshape(rand(norm, 200), (2,100))

m = fit(MvNormal, x)

Z = [pdf(m,[i,j]) for i in -10:10, j in -10:10]
plot(-10:10,-10:10,Z,st=:surface)
```

### 2.3.6 Bayesian inference for the Gaussian

```julia
Σ = ones(2,2)
Σ[1,1] = 3
Σ[2,2] = 17
Σ[1,2] = 7
Σ[2,1] = 7
μ = zeros(2)
d = MvNormal(μ, Σ)
x = rand(d, 1000)

plot(x[1,:],x[2,:],seriestype=:scatter,xlim=(-15,15),ylim=(-15,15))
```

```julia
a0 = -0.3
a1 = 0.5

u = Uniform(-1,1)
x = rand(u,100)
println(size(x))

n = Normal(0,0.04)
ϵ = rand(n,100)

t = a0 .+ a1.*x .+ ϵ

plot(x, a0 .+ a1.*x)
plot!(x, t, seriestype=:scatter)
```

```julia
n = 1000
d = Normal(1.4, 2.7)
x = rand(d, n)
mu = zeros(n)
sigma = zeros(n)
    
mu[1] = (10000*x[1] + (2.7^2)*0)/(10000+2.7^2)
sigma[1] = (10000*2.7^2)/(10000+2.7^2)

for i in 2:n
    mu[i] = ( sigma[i-1]*x[i] + (2.7^2)*mu[i-1] )/(sigma[i-1]+2.7^2)
    sigma[i] = ( sigma[i-1]*2.7^2)/(sigma[i-1]+2.7^2)
end

sigma
```

```julia
plot(Normal(mu[3],sigma[3]))
```
